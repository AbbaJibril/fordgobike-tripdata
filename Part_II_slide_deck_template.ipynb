## Investigation Overview
In this investigation, I want to take a look at how the duration of trips is affected by user type, days of the week, hours of the day, and gender.


## Dataset Overview

The data consists of information regarding approximately 183,000 rides made in a bike-sharing system covering the greater San Francisco Bay area. The data features include duration (secs) and others such as DateTime, customer type, and gender, as well as additional variables such as member's year of birth, start and end station name, start and end station longitude and latitude, start and end station unique id among others. About 8000 data points were removed from the analysis due to inconsistencies or missing information.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('fordgobike-tripdata_clean.csv')
```

## Distribution of bike trips by days
The plot shows that most trips were taken on weekdays, with weekends having the lowest trip records.

```python
# plot a countplot to display the distribution of bike trips across week days

base_color = sb.color_palette()[0]

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekdays= pd.api.types.CategoricalDtype(categories = days, ordered = True)
df['start_weekday'] = df['start_weekday'].astype(weekdays)

plt.figure(figsize = [10, 8])
sb.countplot(data = df, x = 'start_weekday', color = base_color)
plt.title('Distribution of trips by Week-Days')
plt.xticks(rotation = 20)
plt.ylabel('Count of trips Started')
plt.xlabel('Week Day')
```

## Distribution of trips by day hours

The plot show that there are more trips between the 7th and 9th hours, and the 16th and 18th hours. This may be due to people's rush to and from work.

```python
# create a plot count plot to display the distribution of trips across day hours
base_color = sb.color_palette()[0]

plt.figure(figsize = [10, 8])
sb.countplot(data = df, x = 'start_day_time', color = base_color)
plt.title('Distribution of Trips by Day Hours')
plt.ylabel('Count of Trips Started')
plt.xlabel('Day Hour')
```

## Relationship between trip duration and days of the week base on user type and gender.

The plot shows that:
* Customer have more time flexibilty than Subscribers.
* Weekends have more longer duration trips



```python
# plot a lineplot to show the relationship between user types and duration based on their gender across week days
plt.figure(figsize = [15,10])
sb.lineplot(data=df, x="start_weekday", y="duration_min",
             hue="member_gender", style="user_type")
plt.title("Relationship between Trip Duration and Days of the Week based on user tyoe and Gender")
plt.xlabel("Week Day")
plt.ylabel("Duration (min)")
```
